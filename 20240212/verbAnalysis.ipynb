```python
print("Hola!")
with open('sample1.txt') as f:
    # ファイルの内容の読み出し
    data = f.read()  
    data = data.lower()  # 小文字にする
    unitext = data.encode('unicode-escape')

#lay1 = unitext.replace('!','')
#lay2 = lay1.replace('-','')
#lay3 = lay2.replace('\u00a1','')   #¡の削除
#lay4 = lay3.replace('\u00bf','')  #¿の削除
#lay5 = lay4.replace('?','')
#lay6 = lay5.replace('_','')
#lay7 = lay6.replace('.','')
#lay8 = lay7.replace(',','')
#lay9 = lay8.split(' ')
goodbyeList = ['.',',','-','_','\\n',':','\\u00a1','\\u00bf','   ']
Countestar = 0
Countser = 0
lay9 = unitext.decode('utf-8', 'ignore')
lst1 = lay9.split(' ')

for ii in range(len(lst1)):
    for jj in range(len(goodbyeList)):
        string = lst1[ii]
        lst1[ii] = string.replace(goodbyeList[jj],'')
    if lst1[ii] in ['estar' ,'estoy', 'est\\u00e1s','est\u00e1','estamos','est\u00e1is','est\u00e1n',]:
       Countestar +=1
    if lst1[ii] in ['ser', 'soy','eres','es','somos','sois','son']:
        Countser +=1


print(Countestar)
print(Countser)

for ii in range(500):
    print(lst1[ii])
```
